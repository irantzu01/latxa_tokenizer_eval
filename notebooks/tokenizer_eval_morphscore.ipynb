```python
from transformers import AutoTokenizer

model_name = "HiTZ/latxa-7b-v1.1"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
from datasets import load_dataset
import pandas as pd

# load the full dataset (only one split: train)
ds = load_dataset("catherinearnett/morphscore")['train']

# filter by Basque (eus_latn)
df = ds.filter(lambda x: x["language"] == "eus_latn").to_pandas()

print(df.head())
print(len(df))

# save into MorphScoreâ€™s expected folder
df.to_csv("../data/eus_latn_data.csv", index=False)
```

```python
import sys
sys.path.append("/home/irantzu/MASTER/WiSe25/Lab Rotation/morphscore")

from morphscore import MorphScore
morph_score = MorphScore(
    data_dir="../data/",     # <-- this is the important part
    language_subset=['eus_latn'],
    by_split=False,
    freq_scale=True,
    exclude_single_tok=False
)
result, df = morph_score.eval(tokenizer, return_df=True)


metrics = result['eus_latn']   # extract only the metrics dictionary
for key, value in metrics.items():
    print(f"{key}: {value}")
```
